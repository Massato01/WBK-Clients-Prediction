## ***Conecting to PostgreSQL database***

```python
import psycopg2
import pandas as pd
```

### **Conection**

```python
conection = psycopg2.connect(host = 'localhost', database = 'ClientesWBK', user = 'postgres', password = 'VaioMassato201301#', port = 5432)
```

### **Creating cursors for each category**

```python
cursor_clients = conection.cursor()
cursor_credits = conection.cursor()
cursor_jobs = conection.cursor()
cursor_marital_state = conection.cursor()
cursor_fiador = conection.cursor()
cursor_housing = conection.cursor()
cursor_history_credits = conection.cursor()
cursor_investments = conection.cursor()
cursor_otherfinanc = conection.cursor()
cursor_profession = conection.cursor()
cursor_purpose = conection.cursor()

query_clients = 'select * from clientes'
query_credits = 'select * from credito'
query_jobs = 'select * from emprego'
query_marital_state = 'select * from estadocivil'
query_fiador = 'select * from fiador'
query_housing = 'select * from habitacao'
query_history_credits = 'select * from historico_credito'
query_investments = 'select * from investimentos'
query_otherfinanc = 'select * from outrosfinanc'
query_profession = 'select * from profissao'
query_purpose = 'select * from proposito'

cursor_clients.execute(query_clients)
cursor_credits.execute(query_credits)
cursor_jobs.execute(query_jobs)
cursor_marital_state.execute(query_marital_state)
cursor_fiador.execute(query_fiador)
cursor_housing.execute(query_housing)
cursor_history_credits.execute(query_history_credits)
cursor_investments.execute(query_investments)
cursor_otherfinanc.execute(query_otherfinanc)
cursor_profession.execute(query_profession)
cursor_purpose.execute(query_purpose)
```

### **Records**

```python
# Training data
records_clients = cursor_clients.fetchall()
records_credits = cursor_credits.fetchall()

# Testing data
records_jobs = cursor_jobs.fetchall()
records_marital_state = cursor_marital_state.fetchall()
records_fiador = cursor_fiador.fetchall()
records_housing = cursor_housing.fetchall()
records_history_credits = cursor_history_credits.fetchall()
records_investments = cursor_investments.fetchall()
records_otherfinanc = cursor_otherfinanc.fetchall()
records_profession = cursor_profession.fetchall()
records_purpose = cursor_purpose.fetchall()
```

### **Creating the DataFrame**

```python
client = [records_clients[i][1] for i in range(250)]
state = [records_clients[i][2] for i in range(250)]
sex = [records_clients[i][3] for i in range(250)]
duratin = [records_credits[i][0] for i in range(250)]
credit_history = [records_clients[i][1] for i in range(250)]
purpose = [records_credits[i][2] for i in range(250)]
value = [records_credits[i][3] for i in range(250)]
investment = [records_credits[i][4] for i in range(250)]
job = [records_credits[i][5] for i in range(250)]
installment_time = [records_credits[i][6] for i in range(250)] # parcelamento
marital_state = [records_credits[i][7] for i in range(250)]
fiador = [records_credits[i][8] for i in range(250)]
residence_since = [records_credits[i][9] for i in range(250)]
age = [records_credits[i][10] for i in range(250)]
other_finance = [records_credits[i][11] for i in range(250)]
housing = [records_credits[i][12] for i in range(250)]
existing_loan = [records_credits[i][13] for i in range(250)] # emprestimo
profession = [records_credits[i][14] for i in range(250)]
dependents = [records_credits[i][15] for i in range(250)]
socio_company = [records_credits[i][16] for i in range(250)]
foreign = [records_credits[i][17] for i in range(250)]
status = [records_credits[i][18] for i in range(250)]

data = {
    'Client' : client,
    'State' : state,
    'Sex' : sex,
    'Purpose' : purpose,
    'Value' : value,
    'Investment' : investment,
    'Job' : job,
    'InstallmentTime' : installment_time,
    'MaritalState' : marital_state,
    'Fiador' : fiador,
    'EesidenceSince' : residence_since,
    'Age' : age,
    'OtherFinance' : other_finance,
    'Housing' : housing,
    'ExistingLoan' : existing_loan,
    'Profession' : profession,
    'Dependents' : dependents,
    'SocioCompany' : socio_company,
    'Foreign' : foreign,
    'Status' : status
}

df = pd.DataFrame(data)
df
```
